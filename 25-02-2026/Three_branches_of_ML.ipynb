```python
1. Supervised Learning

Definition:
Supervised Learning is a type of machine learning where the dataset contains labels (correct answers).
The model learns a mapping between input â†’ output using these known labels.

Goal:
To predict the correct output for new, unseen data.

Examples:

Classification (Spam or Not Spam)

Regression (House price prediction)
```

```python
2. Unsupervised Learning

Definition:
Unsupervised Learning is a type of machine learning where the dataset does NOT contain labels.
The model tries to find hidden patterns, structures, or clusters in the data.

Goal:
To discover natural groupings or relationships in the data.

Examples:

Clustering

Dimensionality reduction

Pattern discovery
```

```python
3. 3. Reinforcement Learning

Definition:
Reinforcement Learning is a type of machine learning where an agent learns by interacting with an environment and receives rewards or penalties based on its actions.

Goal:
To learn the best action strategy to maximize total reward.

Examples:

Game playing

Robotics

Self-driving systems
```

Comparison Table: Real-World Examples

| Learning Type          | Definition Summary            | Real-World Examples                                       |
| -----------------------| --------------------------    | --------------------------------------------------------  |
| Supervised Learning    | Data has labels               | 1. Spam Email Detection         2. House Price Prediction |
| Unsupervised Learning  | No labels, find patterns      | 1. Customer Segmentation        2. Market Basket Analysis |
| Reinforcement Learning | Learn via rewards & penalties | 1. Game Playing (Chess, Atari)  2. Robot Navigation       |
